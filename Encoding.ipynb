# one hot encoding - in nominal columns

NO ORDER EXIST IN THESE COLUMNS

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder(sparse_output=False,drop='first',handle_unknown='ignore')
# drop = dummy variable trap avoid krta h 
# sparse = avoid sparse matrix
# handle = in future if any new category comes then no error will be shown
```

```python
nominal_cols = ['Street',
    'Neighborhood',
    'Condition1',
    'Condition2',
    'BldgType',
    'HouseStyle',
    'RoofStyle',
    'RoofMatl',
    'Exterior1st',
    'Exterior2nd',
    'MasVnrType',
    'Foundation',
    'Heating',
    
    'Electrical',
    'GarageType',
    
    'SaleType',
    'SaleCondition',
    'LotConfig',
    'LandContour',
    'LandSlope',
    'Utilities',
    'LotShape',
    'MSZoning']
```

```python
encoded_data = ohe.fit_transform(ab[nominal_cols])
#fit - learn catwgories
#transform= perform encoding
#fit_transform = perform both simulateneosly
```

```python
encoded_data
```

```python
encoded_ab = pd.DataFrame(encoded_data,columns=ohe.get_feature_names_out(nominal_cols),index=ab.index)
```

```python
encoded_ab
```

```python
ab = ab.drop(columns=nominal_cols)
```

```python
ab = pd.concat([ab,encoded_ab],axis=1)
```

```python
ab.head()
```

```python
ab_cat_clms
```

```python
ab_cat_clms.columns
```

```python
ordinal_cols = [
    'ExterQual',
    'ExterCond',
    'BsmtQual',
    'BsmtCond',
    'BsmtExposure',
    'BsmtFinType1',
    'BsmtFinType2',
    'HeatingQC',
    'KitchenQual',
    'Functional',
    'FireplaceQu',
    'GarageFinish',
    'GarageQual',
    'GarageCond'
]
```

```python
quality_map = {
    'Po': 1,
    'Fa': 2,
    'TA': 3,
    'Gd': 4,
    'Ex': 5
}

quality_cols = [
    'ExterQual',
    'ExterCond',
    'BsmtQual',
    'BsmtCond',
    'HeatingQC',
    'KitchenQual',
    'FireplaceQu',
    'GarageQual',
    'GarageCond'
]

for col in quality_cols:
    ab[col] = ab[col].map(quality_map)
```

```python
bsmt_exposure_map = {
    'No': 0,
    'Mn': 1,
    'Av': 2,
    'Gd': 3
}

ab['BsmtExposure'] = ab['BsmtExposure'].map(bsmt_exposure_map)
```

```python
bsmt_finish_map = {
    'Unf': 1,
    'LwQ': 2,
    'Rec': 3,
    'BLQ': 4,
    'ALQ': 5,
    'GLQ': 6
}

ab['BsmtFinType1'] = ab['BsmtFinType1'].map(bsmt_finish_map)
ab['BsmtFinType2'] = ab['BsmtFinType2'].map(bsmt_finish_map)
```

```python
garage_finish_map = {
    'Unf': 1,
    'RFn': 2,
    'Fin': 3
}

ab['GarageFinish'] = ab['GarageFinish'].map(garage_finish_map)
```

```python
functional_map = {
    'Sev': 1,
    'Maj2': 2,
    'Maj1': 3,
    'Mod': 4,
    'Min2': 5,
    'Min1': 6,
    'Typ': 7
}

ab['Functional'] = ab['Functional'].map(functional_map)
```

# binary encoding - cantral air


```python
ab.drop(
    columns=['CentralAir_Y', 'PavedDrive_P', 'PavedDrive_Y'],
    inplace=True,
    errors='ignore'
)
```

```python
ab['CentralAir'] = ab['CentralAir'].map({
    'N': 0,
    'Y': 1
})  #binary encoding
```

```python
ab['PavedDrive'] = ab['PavedDrive'].map({
    'N': 0,
    'P': 1,
    'Y': 2
})   #ordinal encoding
```
